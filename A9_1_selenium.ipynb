```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)
total = 22


browser = webdriver.Chrome(service=s)
url = 'https://www.cwa.gov.tw/V8/E/W/County/index.html'
browser.get(url)



```

```python
#getting the city names
for i in range(total):
    cities = browser.find_elements(By.CLASS_NAME, 'city')
    #cities.append(tag2.text)
    #print("\ntag2.tag_name = %s, tag2.text = %s " % (tag2.tag_name, tag2.text))

#for i in range(total):
#    print(cities[i].text)
```

```python
#getting the low temps
for i in range(total):
    temp_low = browser.find_elements(By.XPATH, "//span[@class='tem-C is-active']/i[1]")

#for i in range(total):
#   print(temp_low[i].get_attribute("innerHTML"))

#getting the high temps
for i in range(total):
    temp_high = browser.find_elements(By.XPATH, "//span[@class='tem-C is-active']/i[2]")

#for i in range(total):
#    print(temp_high[i].get_attribute("innerHTML"))
#browser.quit()
```
