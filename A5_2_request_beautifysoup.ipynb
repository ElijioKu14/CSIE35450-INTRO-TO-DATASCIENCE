```python
#beatifulsoup on the website
from bs4 import BeautifulSoup
sp = BeautifulSoup(html.text, 'lxml')

#find and save names of the projects
names = sp.find_all(class_ = "text-md truncate font-mono text-black dark:group-hover/repo:text-yellow-500 group-hover/repo:text-indigo-600 text-smd")
names_ = []
for i in names:
    names_.append(i.text)
    
#print the list of names values
for i in names_:
    print(i)
```

```python

#find and save the last updated time
update = sp.find_all("time")
update_ = []

for i in update:
    update_.append(i.text)
    
#print the list of last updates time values
for i in update_:
    print(i)
```

```python
#find the save the number of downloads
downloads_ = []
temp = sp.find_all(class_="flex-none w-3 text-gray-400 mr-0.5")

for i in temp:
    downloads_.append((i.next_sibling.split()))

#print the list of number of downloads values
for i in downloads_:
    print(i)
```

```python
#find and save the number of likes
likes_ = []
temp1 = sp.find_all(class_="flex-none w-3 text-gray-400 mr-1")
for i in temp1:
    likes_.append(i.nextSibling.split())

#print the list od likes values
for i in likes_:
    print(i)

```
