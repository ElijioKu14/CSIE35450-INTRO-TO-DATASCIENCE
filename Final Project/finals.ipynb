```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup

driverPath = "chromedriver.exe"
s = Service(executable_path=driverPath)
browser = webdriver.Chrome(service=s)

                       
```

```python
url = 'https://www.twitch.tv/directory/category/just-chatting'
browser.get(url)

t1 = []
c1 = []
v1 = []
l1 = []
u1 = []
a1 = []
tn1 = []

titles = browser.find_elements(By.XPATH, "//h3[@class='CoreText-sc-1txzju1-0 MveHm']")
for i in titles:
    t1.append(i.text)
creators = browser.find_elements(By.XPATH, "//p[@class='CoreText-sc-1txzju1-0 gBknDX']")
for i in creators:
    c1.append(i.text)
viewers = browser.find_elements(By.XPATH, "//div[@class='ScMediaCardStatWrapper-sc-anph5i-0 jRUNHm tw-media-card-stat']")
for i in viewers:
    v1.append(i.text)
languages = browser.find_elements(By.XPATH, "//div[@class='Layout-sc-1xcs6mc-0 fAVISI'][1]/div[1]/div[1]/button[1]//div/div/span[1]")
for i in languages:
    l1.append(i.text)
user = browser.find_elements(By.XPATH, "//div[@class='ScTextMargin-sc-10mto54-2 hgouXh']/a[@class='ScCoreLink-sc-16kq0mq-0 eBmhqT preview-card-channel-link tw-link'][1]")
for i in user:
	u1.append(i.get_attribute('href'))
avatar = browser.find_elements(By.XPATH, "//div[@class='ScAvatar-sc-144b42z-0 iYOpTC tw-avatar']/img[@class='InjectLayout-sc-1i43xsx-0 cXFDOs tw-image tw-image-avatar'][1]")
for i in avatar:
	a1.append(i.get_attribute('src'))
thumbnail = browser.find_elements(By.XPATH, "//div[@class='Layout-sc-1xcs6mc-0 dUClxi'][1]//div[@class='Layout-sc-1xcs6mc-0 eFvOkl']//img[@class='tw-image'][1]")
for i in thumbnail:
	tn1.append(i.get_attribute('src'))


```

```python
import pandas as pd
import numpy as np

dict1 = {'Title': t1, 'Creator': c1, 'Viewers': v1, 'Language':l1, 'User Link': u1, 'Avatar Link': a1, 'Thumbnail Link': tn1} 
df1 = pd.DataFrame(dict1)
df1["Viewers"].replace("viewers", '', regex=True, inplace=True)
df1["Viewers"].replace("viewer", '', regex=True, inplace=True)
df1["Viewers"].replace("K", '000', regex=True, inplace=True)
for i in df1["Viewers"]:
    for j in i:
        if(j == "."):
            df1["Viewers"].replace(i,str(int(float(i)*1000)), inplace=True)
df1["Viewers"] = (df1["Viewers"]).astype(int)


dict2 = {'Title': t2, 'Creator': c2, 'Viewers': v2, 'Language':l2, 'User Link': u2, 'Avatar Link': a2, 'Thumbnail Link': tn2} 
df2 = pd.DataFrame(dict2)
df2["Viewers"].replace("viewers", '', regex=True, inplace=True)
df2["Viewers"].replace("viewer", '', regex=True, inplace=True)
df2["Viewers"].replace("K", '000', regex=True, inplace=True)
for i in df2["Viewers"]:
    for j in i:
        if(j == "."):
            df2["Viewers"].replace(i,str(int(float(i)*1000)), inplace=True)
df2["Viewers"] = (df2["Viewers"]).astype(int)

dict3 = {'Title': t3, 'Creator': c3, 'Viewers': v3, 'Language':l3, 'User Link': u3, 'Avatar Link': a3, 'Thumbnail Link': tn3} 
df3 = pd.DataFrame(dict3)
df3["Viewers"].replace("viewers", '', regex=True, inplace=True)
df3["Viewers"].replace("viewer", '', regex=True, inplace=True)
df3["Viewers"].replace("K", '000', regex=True, inplace=True)
for i in df3["Viewers"]:
    for j in i:
        if(j == "."):
            df3["Viewers"].replace(i,str(int(float(i)*1000)), inplace=True)
df3["Viewers"] = (df3["Viewers"]).astype(int)

dict4 = {'Title': t4, 'Creator': c4, 'Viewers': v4, 'Language':l4,'User Link': u4, 'Avatar Link': a4, 'Thumbnail Link': tn4} 
df4 = pd.DataFrame(dict4)
df4["Viewers"].replace("viewers", '', regex=True, inplace=True)
df4["Viewers"].replace("viewer", '', regex=True, inplace=True)
df4["Viewers"].replace("K", '000', regex=True, inplace=True)
for i in df4["Viewers"]:
    for j in i:
        if(j == "."):
            df4["Viewers"].replace(i,str(int(float(i)*1000)), inplace=True)
df4["Viewers"] = (df4["Viewers"]).astype(int)

dict5 = {'Title': t5, 'Creator': c5, 'Viewers': v5, 'Language':l5, 'User Link': u5, 'Avatar Link': a5, 'Thumbnail Link': tn5} 
df5 = pd.DataFrame(dict5)
df5["Viewers"].replace("viewers", '', regex=True, inplace=True)
df5["Viewers"].replace("viewer", '', regex=True, inplace=True)
df5["Viewers"].replace("K", '000', regex=True, inplace=True)
for i in df5["Viewers"]:
    for j in i:
        if(j == "."):
            df5["Viewers"].replace(i,str(int(float(i)*1000)), inplace=True)
df5["Viewers"] = (df5["Viewers"]).astype(int)
```

```python
import matplotlib.pyplot as plt

sum1 = df1["Viewers"].sum()
sum2 = df2["Viewers"].sum()
sum3 = df3["Viewers"].sum()
sum4 = df4["Viewers"].sum()
sum5 = df5["Viewers"].sum()
x = ["Just Chatting", "League of Legends", "Valorant", "GTA5", "Apex Legends"]
y = [sum1,sum2,sum3,sum4,sum5]
plt.bar(x,y)
plt.xticks(rotation=90)
plt.title("The Number of Viewers for each Category")
plt.xlabel("Category")
plt.ylabel("Number of Viewers")
plt.savefig("charts/BarChart.jpg",bbox_inches = 'tight')
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots()

size = 0.3
vals = np.array([df1["Viewers"],df2["Viewers"],df3["Viewers"],df4["Viewers"],df5["Viewers"]])
x = ["Just Chatting", "League of Legends", "Valorant", "GTA5", "Apex Legends"]

cmap = plt.colormaps["tab20c"]
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap([1,2,3,4,5])

ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors, labels=x,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
       wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title="Viewers Distibution")
plt.savefig("charts/viewers_distribution.jpg",bbox_inches = 'tight')
plt.show()
```
