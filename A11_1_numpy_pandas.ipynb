```python
import pandas as pd
import numpy as np


df = pd.read_csv("jOFTZuDrFA9eKZkikfMAwwKEnezbYMxKVW4K3xJP.csv", encoding="utf8")
#df.sort_values('High school system',inplace=True)
df = df.groupby(["High school system","Gender"], as_index=False).sum()
df
# pc 1008
#m 2321
```

```python
'''- Num of graduates
- Num of enroll in police university 
- Num of enroll police college 
- average num of enroll in military school (Round to 2 decimal places) 
- Police university rate (Round to 2 decimal places) 
- Police college rate (Round to 2 decimal places)
- Military school enrollment rate (Round to 2 decimal places) '''
#national = df.iloc[0:12]
#national.drop('High school system', inplace=True, axis=1)
#national = national.groupby("Gender", as_index=False).sum()
#national
h_s_s = df.loc[:,'High school system']
g = df.loc[:,'Gender']

Num_of_grad = df.loc[:,'Num of graduates']
#print(Num_of_grad)

Num_of_enroll_police_uni = df.loc[:,'Num of enroll in police university']
#print(Num_of_enroll_police_uni)

Num_of_enroll_police_college = df.loc[:,'Num of enroll in police college']
#print(Num_of_enroll_police_college)

average_num_of_enroll_military_school = df.loc[:,'Num of enroll in military school']
#print(average_num_of_enroll_military_school)
for i in range(len(average_num_of_enroll_military_school)):
    if (i == 0) or (i == 1):
        average_num_of_enroll_military_school[i] = (average_num_of_enroll_military_school[i]/(len(average_num_of_enroll_military_school)-0)).round(decimals=2)
    elif i == 2 or i == 3:
        average_num_of_enroll_military_school[i] = (average_num_of_enroll_military_school[i]/(len(average_num_of_enroll_military_school)-1)).round(decimals=2)
    elif i == 4 or i == 5:
        average_num_of_enroll_military_school[i] = (average_num_of_enroll_military_school[i]/(len(average_num_of_enroll_military_school)-2)).round(decimals=2)

police_uni_rate = df.loc[:,'Num of enroll in police university']
#print(police_uni_rate)
police_uni_rate = ((police_uni_rate/Num_of_grad)*100).round(decimals=2)
for i in range(len(police_uni_rate)):
    police_uni_rate[i] = str(police_uni_rate[i]) + '%'

police_college_rate = df.loc[:,'Num of enroll in police college']
#print(police_college_rate)
police_college_rate = (police_college_rate/Num_of_grad)*100
police_college_rate = police_college_rate.round(decimals=2)
for i in range(len(police_college_rate)):
    police_college_rate[i] = str(police_college_rate[i]) + '%'

Military_school_enrollment_rate = df.loc[:,'Num of enroll in military school']
#print(Military_school_enrollment_rate)
Military_school_enrollment_rate = (Military_school_enrollment_rate/Num_of_grad)*100
Military_school_enrollment_rate = Military_school_enrollment_rate.round(decimals=2)
for i in range(len(Military_school_enrollment_rate)):
    Military_school_enrollment_rate[i] = str(Military_school_enrollment_rate[i]) + '%'

```

```python
frame = {'High school system': h_s_s,
         'Gender': g,
         'Num of graduates': Num_of_grad,
         'enroll in police academy': Num_of_enroll_police_uni,
         'enroll in police college': Num_of_enroll_police_college,
         'average enroll in military': average_num_of_enroll_military_school,
         'police university enrollment rate': police_uni_rate,
         'police college enrollment rate': police_college_rate,
         'military school enrollment rate': Military_school_enrollment_rate}
 
# Creating DataFrame by passing Dictionary
result = pd.DataFrame(frame)
result
```
